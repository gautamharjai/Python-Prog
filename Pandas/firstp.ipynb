1. Introduction to pandas

What is pandas and Why Use It?


pandas is an open-source Python library that provides high-performance, easy-to-use data structures and data analysis tools.
It is primarily used for manipulating and analyzing data, much like working with tables in Excel or SQL.
Why use pandas?
It provides fast, flexible, and expressive data structures like Series and DataFrame.
pandas excels in handling structured data, such as time series data, tabular data, and heterogeneous data.
It simplifies common data manipulation tasks like reading/writing files, reshaping, merging, and filtering data.
pandas vs Excel: Key Differences
pandas operates programmatically, whereas Excel relies on a graphical interface.
pandas can handle much larger datasets than Excel.
Tasks like filtering, grouping, and analyzing data can be automated with pandas.
pandas supports better integration with other Python libraries for advanced analytics and machine learning.

Data Structures in pandas: Series and DataFrame
Series:

A Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floats, Python objects, etc.).
It can be thought of as a column in an Excel spreadsheet or a SQL table.
Creating a Series:

python
Copy code
import pandas as pd

# Creating a Series from a list
s = pd.Series([1, 3, 5, 7, 9])
print(s)

# Creating a Series with custom indices
s = pd.Series([1, 3, 5, 7, 9], index=['a', 'b', 'c', 'd', 'e'])
print(s)
Key Features of Series:

Indexing: Series can be indexed by labels (custom indices) or by default integers.
Homogeneous data: All elements in a Series are of the same data type.
DataFrame:

A DataFrame is a 2-dimensional labeled data structure, like a table in a relational database or an Excel spreadsheet.
It consists of rows and columns, each of which can have different data types (numeric, string, boolean, etc.).
Creating a DataFrame:

python
Copy code
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 24, 35, 32],
    'City': ['New York', 'Paris', 'Berlin', 'London']
}

df = pd.DataFrame(data)
print(df)
Key Features of DataFrame:

The columns can be of different data types.
It is mutable, meaning you can change data, labels, and even structure.
DataFrames have built-in functions to handle missing data, merge data, filter data, etc.

```python
import pandas as pd

data={'name':["Gautam",'Arjun'],
      'age':[24,33]}
df=pd.DataFrame(data)
print(df)

```
